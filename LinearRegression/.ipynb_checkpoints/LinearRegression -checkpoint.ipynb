# Linear Regression

Linear regression is a fundamental statistical and machine learning technique used to model the relationship between a dependent variable and one or more independent variables. The goal is to find a linear equation that best predicts the dependent variable from the independent variables. 

###  Assumptions of linear regression:

- The Independent variables should be linearly related to the dependent variables.
- Every feature in the data is Normally Distributed.
- There should be little or no multi-collinearity in the data.
- The mean of the residual is zero.
- Residuals obatined should be normally distributed.
- Variance of the residual throughout the data should be same.This is known as homoscedasticity.
- There should be little or no Auto-Correlation is the data.




### There are two types of linear regression:

- Simple Linear Regression:  Involves only one independent variable to predict the dependent variable. It models the relationship between two variables by fitting a linear equation to observed data.

- Multiple Linear Regression: Uses two or more independent variables to predict the dependent variable. It extends simple linear regression by adding more complexity and ability to handle more relationships between variables.


## Simple Linear Regression example:

We'll generate a synthetic dataset using numpy for demonstration purposes, and then we'll implement simple linear regression both manually and using the scikit-learn library.

### Step 1: Generate the Synthetic Dataset

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a synthetic dataset
np.random.seed(0) # For reproducibility
X = 2.5 * np.random.randn(100) + 1.5   # Array of 100 values with mean = 1.5, stddev = 2.5
res = 0.5 * np.random.randn(100)       # Generate 100 residual terms
Y = 2 + 0.3 * X + res                  # Actual values of Y

# Plot the data
plt.scatter(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Synthetic Dataset')
plt.show()
```

### Step 2: Manual Implementation of Simple Linear Regression

We'll calculate the coefficients ($a_1$ and $a_0$ in formula $Y=a_1 X+a_0$) using:

$$
    a_1 = \frac{\sum^{n}_{i=1} (X_i - \bar{X})(Y_i - \bar{Y})}{\sum^{n}_{i=1} (X_i - \bar{X})^{2}}
$$

And

$$
    a_0 = \bar{Y} - b_1 \bar{Y}
$$
 

```python
# Calculate the mean of X and Y
mean_x = np.mean(X)
mean_y = np.mean(Y)

# Total number of values
n = len(X)

# Using the formula to calculate a1 and a0
numerator = 0
denominator = 0
for i in range(n):
    numerator += (X[i] - mean_x) * (Y[i] - mean_y)
    denominator += (X[i] - mean_x) ** 2
a1 = numerator / denominator
a0 = mean_y - (a1 * mean_x)

# Print the coefficients
print(f"Coefficients: a0 = {a0}, a1 = {a1}")
```

### Step 3: Plotting the Linear Regression Line

```python
# Plotting Values and Regression Line
max_x = np.max(X) 
min_x = np.min(X) 

# Calculating line values x and y
x = np.linspace(min_x, max_x, 1000)
y = a0 + a1 * x

# Plotting Line
plt.plot(x, y, color='#58b970', label='Regression Line')
# Plotting Scatter Points
plt.scatter(X, Y, c='#ef5423', label='Scatter Plot')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

### Now use  scikit-learn to Perform Simple Linear Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Reshape X and Y for scikit-learn
X = X.reshape(-1, 1)
Y = Y.reshape(-1, 1)

# Splitting dataset into training and testing set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=0)

# Model initialization
regression_model = LinearRegression()

# Fit the data(train the model)
regression_model.fit(X_train, Y_train)

# Predict
Y_predicted = regression_model.predict(X_test)

# Model Evaluation
mse = mean_squared_error(Y_test, Y_predicted)
print(f"Mean Squared Error: {mse}")

# Plotting Regression Line with scikit-learn
plt.scatter(X_test, Y_test, color='red', label='Test Data')
plt.plot(X_test, Y_predicted, color='green', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

```

Now using Sklearn to make make predictions with a linear regression model, you need to follow a sequence of steps: training the model on your dataset, and then using the trained model to predict the output for new input values.

### Step 1: Train the Model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Assuming X and Y are your dataset's features and target variable respectively
# and they're already defined/loaded into your workspace

# Reshape X to be in the right format expected by scikit-learn if it's a single feature
X = X.reshape(-1, 1)

# Splitting dataset into training and testing set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)

# Create a linear regression model instance
model = LinearRegression()

# Fit the model with the training data
model.fit(X_train, Y_train)
```

### Step 2: Define a Prediction Function

```python
def predict_output(new_input):
    """
    Predicts the output using the trained linear regression model.
    
    Parameters:
    - new_input: A numpy array or a list of new input values for which to predict the output.
    
    Returns:
    - Predicted output for the given input.
    """
    new_input = np.array(new_input).reshape(-1, 1)  # Reshape in case it's a single value or a list
    predicted_output = model.predict(new_input)    # Use the model to predict the output
    return predicted_output
```

### Step 3: Use the Prediction Function

```python
# Example usage
new_input_value = [9.0]  # Example new input value for which you want to predict the output
predicted_output = predict_output(new_input_value)
print(f"Predicted output for input {new_input_value}: {predicted_output}")
```

## Multiple Linear Regression example:

### Step 1: Generate a Synthetic Dataset


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # For 3D plotting

# Set seed for reproducibility
np.random.seed(42)

# Generate synthetic dataset
X1 = 2.5 * np.random.randn(100) + 1.5   # Array of 100 values with mean = 1.5, stddev = 2.5
X2 = 1.5 * np.random.randn(100) + 1    # Array of 100 values with mean = 1, stddev = 1.5
res = 0.5 * np.random.randn(100)       # Generate 100 residual terms
Y = 2 + 1.1*X1 + 0.5*X2 + res           # Actual values of Y

# Plotting the dataset
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X1, X2, Y, c='r', marker='o')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')
plt.title('3D Scatter plot of Synthetic Dataset')
plt.show()
```

In the generation of a synthetic dataset for a linear regression example, the residual term plays a crucial role in adding realism and complexity to the data. Specifically, the residual term represents the error component or noise in the data that cannot be explained by the linear relationship between the independent variables (X1, X2) and the dependent variable (Y). 

### Step 2: Fit a Multiple Linear Regression Model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Reshaping and combining X1 and X2 into a single feature array
X = np.column_stack((X1, X2))

# Splitting dataset into training and testing set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Creating a linear regression model
model = LinearRegression()

# Fitting the model with the training data
model.fit(X_train, Y_train)

# Predicting Y for X_test
Y_pred = model.predict(X_test)
```

Fitting a multiple linear regression model involves several steps, each crucial for understanding the relationship between the independent variables and the dependent variable:


- Combining Features: Since we have two independent variables (X1 and X2), we combine them into a single feature array X. This is necessary because the LinearRegression model in scikit-learn expects the features to be in a two-dimensional array, where each row is a sample and each column is a feature.

- Splitting the Dataset: We split the dataset into a training set and a testing set using train_test_split. This allows us to train the model on one subset of the data and evaluate its performance on a separate subset, providing insight into how well the model generalizes to unseen data.

- Creating and Training the Model: We instantiate the LinearRegression model and then call its fit method with the training data. This process involves finding the coefficients (weights) for the independent variables that minimize the sum of squared residuals (the differences between the observed and predicted values of Y in the training set).

- Model Coefficients: After training, the model has coef_ attributes representing the learned coefficients for each independent variable and an intercept_ attribute representing the y-intercept. These coefficients quantify the relationship between each independent variable and the dependent variable, adjusting for the presence of other independent variables.



### Step 3: Plotting the Model Predictions

```python
# Creating a meshgrid for X1 and X2
x1_range = np.linspace(X1.min(), X1.max(), 20)
x2_range = np.linspace(X2.min(), X2.max(), 20)
x1_grid, x2_grid = np.meshgrid(x1_range, x2_range)

# Predicting Y for each combination of X1 and X2 on the grid
y_grid = model.predict(np.hstack((x1_grid.reshape(-1, 1), x2_grid.reshape(-1, 1))))

# Plotting the original dataset
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X1, X2, Y, c='r', marker='o', label='Original Data')

# Plotting the plane of predicted values
ax.plot_surface(x1_grid, x2_grid, y_grid.reshape(x1_grid.shape), color='b', alpha=0.3)

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')
plt.title('Multiple Linear Regression Model')
ax.legend()
plt.show()
```
